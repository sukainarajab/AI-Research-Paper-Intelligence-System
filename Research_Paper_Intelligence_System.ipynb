```python
def extract_entities(text):
    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
    return entities
```

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
candidate_labels = ["Computer Vision", "Natural Language Processing", "Reinforcement Learning",
                    "Generative Models", "Optimization", "Graph Neural Networks", "Robotics"]
def classify_paper(text):
    result = classifier(text, candidate_labels)
    return result["labels"][0], result["scores"][0]
```

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
def answer_question(context, question):
    result = qa_pipeline(question=question, context=context)
    return result["answer"], result["score"]
```

```python
def full_paper_analysis(query, k=5, question=None):
    query_embedding = model.encode([query])
    faiss.normalize_L2(query_embedding)
    D, I = index.search(query_embedding, k)

    for score, idx in zip(D[0], I[0]):
        abstract = df.iloc[idx]["abstract"]
        print("Similarity Score :", score)
        print("Title:", df.iloc[idx]["title"])
        print()

        summary = summarizer(abstract, max_length=120, min_length=40, do_sample=False)
        summary_text = summary[0]["summary_text"]
        print("Summary:", summary_text)
        print()

        keywords = kw_model.extract_keywords(abstract, keyphrase_ngram_range=(1, 3), stop_words="english")
        print("Keywords:", keywords)
        print()

        entities = extract_entities(summary_text)
        print("Named Entities:")
        for entity, label in entities:
            print(f"{entity} --> {label}")
        print()

        label, conf = classify_paper(abstract)
        print(f"Predicted Topic: {label} (confidence: {conf:.2f})")

        if question:
            answer, ans_score = answer_question(abstract, question)
            print()
            print(f"Q: {question}")
            print(f"A: {answer} (confidence: {ans_score:.2f})")
        print("-" * 80)
```

```python
full_paper_analysis("deep learning for medical image analysis", k=5,
                     question="What method does the paper use?")
```
